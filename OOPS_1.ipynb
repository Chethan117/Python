```python
class Student():

  def __init__(self,my_name="Unamed",age = 0):#Parameterized Condtructor, with defult value, if no arg is passed
    self.name = my_name
    self.age=age
```

# Lets add roll number to object and see

```python
class Student():
 # constructor
 def __init__(self, my_name,age):
   self.name = my_name
   self.age = age
   self.counter = 0
   self.counter+=1
   self.rollno=self.counter

 # representation magic method
 def __repr__(self):
   return "Student is {}.".format(self.name)
 
 def introduce(self):
   return f"Hello my name is {self.name}"
```

```python
s1 = Student("John", 15)
s2 = Student("Bill", 17)
s3 = Student("Don", 14)
```

```python
print(s1.rollno)
print(s2.rollno)
print(s3.rollno)
```

## Why all s1, s2, s3 giving same roll no, because every time u create new obect counter will be initialized to 0 inside constructor, that means contructor is newly called for each object

```python
class Student():

 # class variable
 counter = 0

 # constructor
 def __init__(self, my_name, age):
  Student.counter += 1

 # instance variables
  self.name = my_name
  self.age = age
  self.rollno = Student.counter

 # representation magic method
 def __repr__(self):
  return "Student is {}.".format(self.name)

 # instance method
 def introduce(self):
  return f"Hello! My name is {self.name}"
```

```python
s1 = Student("John", 15)
s2 = Student("Bill", 17)
s3 = Student("Don", 14)
```

```python
print(s1.rollno)
print(s2.rollno)
print(s3.rollno)
```
