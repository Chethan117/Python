```python
class Account():
 counter = 1000

 def __init__(self, balance = 0):
  Account.counter +=1
  self.accnum = Account.counter
  self._balance = balance
  self.passbook= []


 def __repr__(self):
  return f"Account {self.accnum} has balance {self._balance}."
 
 def deposit(self, amount):
  if amount>0:
   self._balance += amount
   entry=('credit', amount)
   self.passbook.append(entry)
  
 

 def withdrawl(self, amount):
  if amount>0 and amount<= self._balance:
   self._balance -= amount
   entry=('debit', amount)
   self.passbook.append(entry)


class SavingsAccount(Account):
  def getInterest(self):
    return self._balance*0.05
 
class CurrentAccount(Account):
  def __init__(self):
    super().__init__() ##INHERITING  INIT from parent class  
    #Account.__init__(self) Instead of  super().__init__() , u can write like this as wel;

    self.maxTrans = 5

  def getInterest(self):
    return self._balance*0.01



sa1 = SavingsAccount()
ca1 = CurrentAccount()
print(sa1)
sa1.deposit(100)
sa1.deposit(200)
print(sa1)
sa1.withdrawl(50)
print(sa1)
print("-"*50)
print(ca1)
ca1.deposit(500)
ca1.deposit(300)
print(ca1)
ca1.withdrawl(200)
print(ca1)
```
