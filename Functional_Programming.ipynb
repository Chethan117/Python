# Pass by reference

```python
lst = [0,1,2]
def modify_content(a):
 a[0]= 999
 return a
print(lst)
print(modify_content(lst))
print(lst)
```

# Pass by value or assigning value

```python
lst = [0,1,2]
def modify_content(a):
 a = [4,5,6]
 return a
print(lst)
print(modify_content(lst))
print(lst)
```

```python
st = "Hello world"
def modify_content(a):
 a[0]= "Y"
 return a
print(st)
print(modify_content(st))
print(st)
```

```python
heights = [152, 182, 170, 161, 180, 172, 146]
```

<150 : S
150-170 : M
180 : L


```python
def height_func(h):
 if h < 150:
   return 'S'
 elif h >=150 and h < 170:
   return 'M'
 return 'L'
```

```python
list(map(height_func,heights))
```

```python
A = list(range(1,11))
B = reversed(A)
reduce(lambda x,y:x*y, A) == reduce(lambda a,b:a*b,B)
```
