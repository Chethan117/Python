```python
def insertion_sort(arr):
  for i in range(1,len(arr)):
    key = arr[i]
    j = i - 1

    while j>=0 and arr[j] > key:
      arr[j+1] = arr[j]
      j-=1
      arr[j+1] = key

  return arr
  
```

```python
arr = [5,8,0,2,1]
insertion_sort(arr)
```
